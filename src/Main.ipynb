### Creating Cohorts based on Tenure

**Let's begin by treating each unique tenure length, 1 month, 2 month, 3 month...N months as its own cohort.**

**Treating each unique tenure group as a cohort, calculating the Churn rate (percentage that had Yes Churn) per cohort. For example, the cohort that has had a tenure of 1 month have a Churn rate of 61.99%. we should have cohorts 1-72 months with a general trend of the longer the tenure of the cohort, the less of a churn rate. This makes sense as we are less likely to stop service the longer you've had it.**

```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
```

```python
churn_rate = 100 * yes_churn / (no_churn+yes_churn)
```

```python
churn_rate.transpose()['customerID']
```

**Now that we have Churn Rate per tenure group 1-72 months, Let's create a plot showing churn rate per months of tenure.**

```python
plt.figure(figsize=(10,4),dpi=200)
churn_rate.iloc[0].plot()
plt.ylabel('Churn Percentage');
```

### Broader Cohort Groups
**Based on the tenure column values, create a new column called Tenure Cohort that creates 4 separate categories:**
   * '0-12 Months'
   * '24-48 Months'
   * '12-24 Months'
   * 'Over 48 Months'    

```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return "Over 48 Months"
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df.head(10)[['tenure','Tenure Cohort']]
```

**Creating a scatterplot of Total Charges versus Monthly Charts,colored by Tenure Cohort defined in the previous task.**

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(data=df,x='MonthlyCharges',y='TotalCharges',hue='Tenure Cohort', linewidth=0.5,alpha=0.5,palette='Dark2')
```

**Creating a count plot showing the churn count per cohort.**

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(data=df,x='Tenure Cohort',hue='Churn')
```

# Part 4: Predictive Modeling

**Let's explore 4 different tree based methods: A Single Decision Tree, Random Forest, AdaBoost, Gradient Boosting.**


## Single Decision Tree

**Separating out the data into X features and Y label. Create dummy variables where necessary and features which are not useful and should be dropped.**

```python
X = df.drop(['Churn','customerID'],axis=1)
X = pd.get_dummies(X,drop_first=True)
```

```python
y = df['Churn']
```

**Performing a train test split, holding out 10% of the data for testing. We'll use a random_state of 101 in the solutions notebook/video.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=6)
```

```python
dt.fit(X_train,y_train)
```

```python
preds = dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,classification_report,confusion_matrix
```

```python
print(classification_report(y_test,preds))
```

```python
cm=confusion_matrix(y_test,preds)

ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=dt.classes_).plot()
```

```python
imp_feats = pd.DataFrame(data=dt.feature_importances_,index=X.columns,columns=['Feature Importance']).sort_values("Feature Importance")
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.barplot(data=imp_feats.sort_values('Feature Importance'),x=imp_feats.sort_values('Feature Importance').index,y='Feature Importance')
plt.xticks(rotation=90)
plt.title("Feature Importance for Decision Tree");
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(dt,filled=True,feature_names=X.columns);
```

## Random Forest

```python
rf.fit(X_train,y_train)
```

```python
preds = rf.predict(X_test)
```

```python
cm=confusion_matrix(y_test,preds)

ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=rf.classes_).plot()
```
